```python
def unify(s1, s2, theta=None):
    if theta is None:
        theta = {}

    if s1 == s2:
        return theta
    elif isinstance(s1, str) and s1.islower():
        return unify_var(s1, s2, theta)
    elif isinstance(s2, str) and s2.islower():
        return unify_var(s2, s1, theta)
    elif isinstance(s1, list) and isinstance(s2, list) and len(s1) == len(s2):
        theta = unify(s1[0], s2[0], theta)
        if theta is False:
            return False
        return unify(s1[1:], s2[1:], theta)
    else:
        return False

def unify_var(var, x, theta):

    if var in theta:
        return unify(theta[var], x, theta)
    elif x in theta:
        return unify(var, theta[x], theta)
    elif occurs_check(var, x, theta):
        return False
    else:
        theta[var] = x
        return theta

def occurs_check(var, x, theta):
    if var == x:
        return True
    elif isinstance(x, str) and x in theta:
        return occurs_check(var, theta[x], theta)
    elif isinstance(x, list):
        for item in x:
            if occurs_check(var, item, theta):
                return True
    return False

def substitute(theta, expression):
    if isinstance(expression, str):
        return theta.get(expression, expression)
    elif isinstance(expression, list):
        return [substitute(theta, item) for item in expression]
    else:
        return expression
```

```python

s1 = ['likes', 'Ram', 'y']
s2 = ['likes', 'x', 'Raj']
unification_result1 = unify(s1, s2)
print(f"Unification for {s1} and {s2}: {unification_result1}")

t1 = ['P', 'x', 'y']
t2 = ['Q', 'a', ['f', 'z']]
unification_result2 = unify(t1, t2)
print(f"Unification for {t1} and {t2}: {unification_result2}")
```
